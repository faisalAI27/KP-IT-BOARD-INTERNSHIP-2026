```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
color_image=Image.open('cat_image.jpg').convert('RGB')
print('image size',color_image.size)
print('mode',color_image.mode)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(color_image)
plt.title('original image')
plt.axis('off')
plt.show()
```

```python
gray_image=color_image.convert('L')
```

```python
plt.figure(figsize=(8,6))
plt.imshow(gray_image,cmap='gray')
plt.title('Gray Scale Image')
plt.axis('off')
plt.show()
```

```python
gray_pixel=np.array(gray_image)
print('matrix shape:',gray_pixel.shape)
print(gray_pixel)
```

```python
height,width=gray_pixel.shape
center_y=height//2
center_x=width//2
pixel_block=gray_pixel[center_y-5:center_y+5,
                       center_x-5:center_x+5

]
print(pixel_block)
```

```python
plt.figure(figsize=(8, 8))

plt.imshow(
    pixel_block,
    cmap="gray",
    vmin=0,
    vmax=255
)

for row in range(pixel_block.shape[0]):
    for column in range(pixel_block.shape[1]):
        value = pixel_block[row, column]

        plt.text(
            column,
            row,
            str(value),
            ha="center",
            va="center"
        )

plt.title("10 × 10 Grayscale Pixel Values")
plt.xticks(range(pixel_block.shape[1]))
plt.yticks(range(pixel_block.shape[0]))
plt.show()
```

```python
screenshot_image=Image.open('screenshot_cat.jpg').convert('RGB')
```

```python
print('screenshot size',screenshot_image.size)
print('shape:',screenshot_image.mode)
```

```python
plt.figure(figsize=(8,6))
plt.imshow(screenshot_image)
plt.title('Screenshot Image')
plt.axis('off')
plt.show()
```

```python
screenshot_gray = screenshot_image.convert("L")

plt.figure(figsize=(8, 6))
plt.imshow(screenshot_gray, cmap="gray")
plt.title("Grayscale Screenshot")
plt.axis("off")
plt.show()
```

```python
screenshot_pixels = np.array(screenshot_gray)

print("Screenshot pixel matrix shape:", screenshot_pixels.shape)
print(screenshot_pixels)
```

```python
screenshot_height, screenshot_width = screenshot_pixels.shape

screenshot_center_y = screenshot_height // 2
screenshot_center_x = screenshot_width // 2

screenshot_block = screenshot_pixels[
    screenshot_center_y - 5:screenshot_center_y + 5,
    screenshot_center_x - 5:screenshot_center_x + 5
]

print(screenshot_block)
```

```python
plt.figure(figsize=(8, 8))

plt.imshow(
    screenshot_block,
    cmap="gray",
    vmin=0,
    vmax=255
)

for row in range(screenshot_block.shape[0]):
    for column in range(screenshot_block.shape[1]):
        value = screenshot_block[row, column]

        plt.text(
            column,
            row,
            str(value),
            ha="center",
            va="center"
        )

plt.title("10 × 10 Screenshot Pixel Values")
plt.xticks(range(screenshot_block.shape[1]))
plt.yticks(range(screenshot_block.shape[0]))
plt.show()
```
